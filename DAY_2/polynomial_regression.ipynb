```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate data
np.random.seed(0)
X = np.linspace(0, 10, 100).reshape(-1, 1)
y = np.sin(X).ravel() + np.random.normal(scale=0.3, size=X.shape[0])

# Linear Regression model
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Plot
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, y_pred, color='red', label='Linear Fit (Underfitting)')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline

# High-degree polynomial model (degree = 15)
degree = 15
model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
model.fit(X, y)
y_pred = model.predict(X)

# Plot
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, y_pred, color='green', label='15th Degree Polynomial Fit (Overfitting)')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Step 1: Load data from CSV file
data = pd.read_csv('house_data.csv')  # Replace with the correct path if necessary
# Step 2: Extract features (size) and target (price)
sizes = data['size'].values.reshape(-1, 1)  # House sizes
prices = data['price'].values  # House prices
# Step 3: Apply Polynomial Transformation
# Degree 2 polynomial (quadratic)
poly = PolynomialFeatures(degree=2)
sizes_poly = poly.fit_transform(sizes)

# Step 4: Train a Polynomial Regression Model
model = LinearRegression()
model.fit(sizes_poly, prices)
# Step 5: Make predictions (using the same data)
predicted_prices = model.predict(sizes_poly)

# Step 6: Visualize the results
plt.scatter(sizes, prices, color='blue', label='Actual Prices')  # Plot the actual data
plt.plot(sizes, predicted_prices, color='red', label='Polynomial Regression (degree=2)')  # Polynomial curve
plt.xlabel('House Size (Square Feet)')
plt.ylabel('House Price (Thousands of $)')
plt.title('House Size vs Price - Polynomial Regression')
plt.legend()
plt.show()
```
